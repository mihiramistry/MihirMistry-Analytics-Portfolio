TASK 3 – Grouping and Aggregation

```python
#Computing the average Life Expectancy and GDP per capita by country
country_means = (
    df.groupby('Country')[['LifeExpectancy', 'GDPperCapita']]
    .mean()
    .sort_values(by='LifeExpectancy', ascending=False)
)

print("Top 5 countries by average Life Expectancy:")
print(country_means['LifeExpectancy'].head(5))
print("\n")

print("Top 5 countries by average GDP per Capita:")
print(country_means.sort_values(by='GDPperCapita', ascending=False)['GDPperCapita'].head(5))
print("\n")

#Computing the average GDP Growth Rate per decade

# Creating a new column for decade
def get_decade(year):
    if 1990 <= year < 2000:
        return "1990s"
    elif 2000 <= year < 2010:
        return "2000s"
    elif 2010 <= year <= 2020:
        return "2010s"
    else:
        return "Other"

df['Decade'] = df['Year'].apply(get_decade)

# Calculating average GDP Growth by decade
gdp_growth_decade = (
    df.groupby('Decade')['GDPGrowth']
    .mean()
    .sort_index()
)

print("Average GDP Growth Rate by Decade:")
print(gdp_growth_decade)
```

Explanation: Here I used groupby to figure out the average life expectancy and GDP per capita of each country and filtered the best performing countries.
Eventually, I also calculated the average growth of the GDP per decade (1990s, 2000s, 2010s). This also highlight the economic and health trends over time and across nations.

TASK 6 – Creating New Columns

```python
#Creating PopulationMillions column
df['PopulationMillions'] = df['Population'] / 1_000_000

#Calculating GDP thresholds
gdp_median = df['GDPperCapita'].median()
gdp_75th = df['GDPperCapita'].quantile(0.75)

print(f"Median GDP per Capita: {gdp_median:.2f}")
print(f"75th Percentile GDP per Capita: {gdp_75th:.2f}\n")

#Creating GDPCategory column
def categorize_gdp(gdp):
    if gdp < gdp_median:
        return "Low"
    elif gdp_median <= gdp <= gdp_75th:
        return "Medium"
    else:
        return "High"

df['GDPCategory'] = df['GDPperCapita'].apply(categorize_gdp)

#Counting how many rows fall into each GDP category
category_counts = df['GDPCategory'].value_counts()

print("Number of Rows in Each GDP Category:")
display(category_counts)
```

Explanation:The population values should be simplified to millions, GDP category in order to categorize each observation as low, medium or high income.
Lastly, I enumerated the number of rows which belong in each GDP category. This form useful types of groupings with which to compare later.

TASK 10 – Compare 1990 vs 2020

```python
#Identifying countries with data for BOTH 1990 and 2020
df_1990 = df[df['Year'] == 1990][['Country', 'LifeExpectancy']]
df_2020 = df[df['Year'] == 2020][['Country', 'LifeExpectancy']]

#Merging on Country to keep only those present in BOTH years
merged = pd.merge(df_1990, df_2020, on="Country", suffixes=("_1990", "_2020"))

print("Countries with data for BOTH 1990 and 2020:")
display(merged.head())

#Computing change in LifeExpectancy (2020 - 1990)
merged['LifeExp_Change'] = merged['LifeExpectancy_2020'] - merged['LifeExpectancy_1990']

print("\nLife Expectancy Change (1990 → 2020):")
display(merged.head())

#Top 3 countries with greatest improvement
top3_improvement = merged.sort_values(by='LifeExp_Change', ascending=False).head(3)

print("\nTop 3 Countries with the Greatest Improvement:")
display(top3_improvement)
```

Data of the countries for both the year 1990 and 2000 was identified, along with the changes in life expectancy for each year of these countires, out of which top 3 was sorted out which was having maximum improvements, which results in listing down the countries which had the highest progess in last 30 years.
